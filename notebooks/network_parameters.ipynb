```python
mask_net_dir = os.path.join(PROJECT_ROOT, "output",  "training_91_mask_net_prediction_thresholds_spars0.001" )
pre_mask_net_dir = os.path.join(PROJECT_ROOT, "output", "training_74_pre_mask_net_spars0.001")

mask_net_cfg = os.path.join(mask_net_dir, "cfg_mask_net_thresholds_train.yml")
pre_mask_net_cfg = os.path.join(pre_mask_net_dir, "cfg_pre_mask_net.yml")
```

```python
def get_setup(config):
    argv  = ["-c", config]
    
    return SetupDiagnostics(argv)
```

```python
mask_net_setup = get_setup(mask_net_cfg)
pre_mask_net_setup = get_setup(pre_mask_net_cfg)
```

```python
mask_net_models = load_models(mask_net_setup)
pre_mask_net_model = load_models(pre_mask_net_setup)
```

```python
mask_net_models = {var: values[0] for var, values in mask_net_models[mask_net_setup.nn_type].items()}
pre_mask_net_model = {var: values[0] for var, values in pre_mask_net_model[pre_mask_net_setup.nn_type].items()}
```

```python
def print_params(models):
    for var, model in models.items():
        num_parameters = np.sum([np.prod(v.shape) for v in model.trainable_variables])        
        print(f"Variable {var}: {num_parameters}")
```

```python
print("PreMaskNet Params\n")
print_params(pre_mask_net_model)
```

```python
print("MaskNet Params\n")
print_params(mask_net_models)
```
